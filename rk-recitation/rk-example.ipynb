```python
from numpy import pi, linspace
import matplotlib.pyplot as plt
from pendulum import solve_pendulum

# Solve for a specific a, b
alpha   = 1
beta    = 100
u       = pi/4
v       = 30
T       = 5
h       = 1e-2
t       = linspace(0,T,int(T/h) + 1)
U,V     = solve_pendulum(alpha,beta,T,u,v,h)

# Plotting Stuff
plt.plot(t,U)
```
